# Operación Forum

Se trata de un CCS CLFCLP Fija vs Fija a 3Y donde el Banco recibe CLP y paga CLF.

## Librerías

```python
import qcfinancial as qcf
import aux_functions as aux
```

## Características Comunes de Ambas Patas

```python
str_maturity = "3Y"
maturity = qcf.Tenor(str_maturity)
meses = maturity.get_months() + 12 * maturity.get_years()
```

```python
common_parameters = {
    "settlement_periodicity": qcf.Tenor('6M'),
    "start_date": (start_date:=qcf.QCDate(13, 10, 2025)),
    "end_date": start_date.add_months(meses),
    "bus_adj_rule": qcf.BusyAdjRules.MODFOLLOW,
    "settlement_calendar": qcf.BusinessCalendar(start_date, 5),
    "settlement_stub_period": qcf.StubPeriod.NO,
    "settlement_lag": 0,
    "amort_is_cashflow": True,
    "sett_lag_behaviour": qcf.SettLagBehaviour.DONT_MOVE,
    "is_bond":False,
}
```

## Pata Fija en CLF (Banco recibe)

```python
fixed_rate_value = .0356
clf_leg_other_values = {
    "rec_pay": qcf.RecPay.PAY,
    "initial_notional": 1_000_000.0,
    "notional_currency": qcf.QCCLF(),
    "interest_rate": qcf.QCInterestRate(fixed_rate_value, qcf.QCAct360(), qcf.QCLinearWf()),
}
```

```python
pata_clf = qcf.LegFactory.build_bullet_fixed_rate_leg(
     **(common_parameters | clf_leg_other_values)
)
```

```python
aux.leg_as_dataframe(pata_clf)
```

## Pata Fija en CLP (Banco paga)

```python
fixed_rate_value = .07
clp_leg_other_values = {
    "rec_pay": qcf.RecPay.RECEIVE,
    "initial_notional": 39_505_990_000.0,
    "notional_currency": qcf.QCCLP(),
    "interest_rate": qcf.QCInterestRate(fixed_rate_value, qcf.QCAct360(), qcf.QCLinearWf()),
}
```

```python
pata_clp = qcf.LegFactory.build_bullet_fixed_rate_leg(
     **(common_parameters | clp_leg_other_values)
)
```

```python
aux.leg_as_dataframe(pata_clp)
```

```python
rec = clf_leg_other_values["initial_notional"] * .05 * 39_505.99
print(f"REC: {rec:,.0f} CLP")
```
